# 逻辑

## 1. 命题

能够判断对错的陈述句叫作命题 (Proposition), 例如如下陈述句都是命题:

- 5 + 5 = 10, 结果为 "对"
- Alice (13 岁) 的年龄在 6 岁以上, 结果为 "对"
- 5 + 5 = 11, 结果为 "错"
- Bob (6 岁) 的年龄在 6 岁以下, 结果为 "错"

命题逻辑是逻辑学中一个最基本的概念, 它定义了逻辑系统的基本元素. 命题是逻辑系统的基本单元，命题的取值为 "真" 或 "假"

命题正确时, 称该命题为 "真", 反之，命题不正确时，称该命题为 "假". "真" 称作 `true`, "假" 称作 `false`

- 5 + 5 = 10 => `true`
- Alice (13 岁) 的年龄在 6 岁以上 => `true`
- 5 + 5 = 11 => `false`
- Bob (6 岁) 的年龄在 6 岁以下 => `false`

命题要么为 `true` 要么为 `false`, 同时满足 `true` 和 `false` 的不能称为命题; 既不为 `true` 也不为 `false` 的也不能称为命题

### 1.1. 真值表

真值表是一个表格，用于记录命题的真值, 列数表示命题的取值，行数表示命题的真值情况

对于一个命题 `A`, 其结果一定为 `true` 或 `false`, 故其指针表如下:

| A     |
| ----- |
| True  |
| False |

## 2. 规则

所谓规则, 指的是一个具备自变量的命题, 当自变量的值导致命题为 `true` 或 `false` 时, 能够各自对应一个因变量, 例如一个乘车买票的规则:

- 当乘客年龄大于等于 6 岁时, 票价 100 元;
- 当乘客年龄小于 6 岁时, 票价 0 元;

其中的自变量是乘客的年龄, 因变量是票价, 因变量的计算方式是自变量是否导致命题为 `true` 或 `false`

### 2.1. 规则遗漏

一个规则应该覆盖所有可能的因变量值, 否则就会发送规则遗漏, 也就是存在至少一个自变量值, 令规则的命题既不为 `true` 也不为 `false`, 从而无法计算出因变量, 例如:

- 当乘客年龄大于 6 岁时, 票价 100 元;
- 当乘客年龄小于 6 岁时, 票价 0 元;

上述规则在 "乘客年龄等于 6 岁时", 命题既不为 `true` 也不为 `false`, 无法计算出因变量, 存在规则遗漏

### 2.2. 规则重复

所谓规则重复, 是指存在至少一个自变量, 令规则的可以同时为 `true` 和 `false`, 即规则重复, 例如:

- 当乘客年龄小于等于 6 岁时, 票价 0 元;
- 当乘客年龄大于等于 6 岁时, 票价 100 元;

当乘客年龄是 6 岁时, 上述规则产生了两个结果, 且两个结果自相矛盾, 存在规则重复

### 2.3. 完整性和排他性

在考虑规则时，确认有没有 "遗漏" 和 "重复" 是相当重要的:

- 没有 "遗漏"，即具备完整性，表示该规则在任何情况下都适用;
- 没有 "重复"，即具备排他性，表示该规则在任何情况下都不会产生矛盾;

在遇到大问题时，通常将其分解为多个小问题, 这时常用的方法就是检查规则的完整性和排他性. 即使是难以解决的大问题，也能通过这种方法转换成容易解决的小问题. 这种方法也称为 MECE (Mutually Exclusive and Collectively Exhaustive)

### 2.4. 文氏图

文氏图 (也称为韦恩图或者范氏图), 原本是用于展示集合关系的一种图, 也可以直观的用于规则的完整性或排他性检查, 例如:

![X](./.assets/venn-diagram.png)

上面的图显示了两个范围, 矩形框为所有值的全集, 圆形框内为 "小于等于 6 岁时的值集合", 矩形和圆形相差的部分为 "6 岁以上的值集合"

## 3. 复杂命题

由多个子命题共同组成的命题，称为复杂命题, 复杂命题是通过多个简单命题, 通过逻辑运算符连接起来得到的

逻辑运算符:

- 逻辑非: `¬`
- 逻辑与: `∧`
- 逻辑或: `∨`
- 逻辑异或: `⊕`

### 3.1. 逻辑非

逻辑非会对一个命题结果取反, 即对于命题 `A`, 其逻辑非运算表达式为 $\neg A$, 表示 `not A`

#### 3.1.1. 真值表

下面表示了命题 `A` 及其**非**逻辑运算结果的真值表, 可以无歧义的表达非逻辑运算的各种结果

| A     | ¬A    |
| ----- | ----- |
| True  | False |
| False | True  |

该真值表为我们展现了运算符 `¬` 的定义，具体为:

- 命题 `A` 为 `true` 时，命题 `¬A` 为 `false`
- 命题 `A` 为 `false` 时，命题 `¬A` 为 `true`

因为 `A` 是命题，所以它要么是 `true`，要么是 `false`. 因此，该真值表覆盖了所有情况, 没有遗漏和重复，兼顾了完整性和排他性


#### 3.1.2. 双重否定

对于命题 `A`，否定即逻辑非运算 `¬A`, 双重否定即叠加两重逻辑非运算, 即 `¬¬A`, 双重否定等于肯定, 故 `¬¬A` 等于 `A`

双重否定的真值表如下:

| A     | ¬A    | ¬¬A   |
| ----- | ----- | ----- |
| True  | False | True  |
| False | True  | False |

#### 3.1.3. 文氏图

文氏图是一种用于表示逻辑关系的图形，通常用圆圈表示变量，用线表示逻辑关系, 例如对于命题 `A`, 其文氏图为:

![X](./.assets/venn-diagram-for-not.png)

文氏图原本是表示集合关系的图, 外围的矩形表示全集, 这里则表示 "一星期所有日子的集合", 假设 `A` 为命题 "乘车日是星期日", 则矩形内部的椭圆表示 "星期日的集合", 也就是说，该区域表示 "命题 `A` 为 `true` 的日子的集合"

那么，如果矩形区域是 "一星期所有日子的集合"，而椭圆内的区域是 "星期日的集合" 的话，椭圆以外的部分即为 "不是星期日的日子的集合". 这里称为 "命题 `A` 为 `false` 的日子的集合" 或者 "命题 `¬A` 为 `true` 的日子的集合"

### 3.2. 逻辑 "与"

对于命题 `A` 和 `B`，仅当两个命题都为 `True` 时, 命题 `A` 与 `B` 的 "与" 运算结果才为 `True`

两个命题的 "与" 运算可写为: $A \wedge B$, 表示 `A and B`

#### 3.2.1. 逻辑 "与" 真值表

基于 "与" 运算的规则, 两个命题的 "与" 运算真值表如下:

| A     | B     | A ∧ B |
| ----- | ----- | ----- |
| True  | True  | True  |
| True  | False | False |
| False | True  | False |
| False | False | False |

从表中可以看出, 对于 `A` 和 `B` 两个命题可组合的 4 中情况中, 当且仅当`A` 和 `B` 同时为 "真" 时, $A \wedge B$ 才为 "真"

#### 3.2.2. 逻辑 "与" 文氏图

对于 `A` 和 `B` 两个命题, 其逻辑 "与" 运算的文氏图如下:

![X](./.assets/venn-diagram-for-and.png)

两个椭圆分别表示 `A` 和 `B` 命题, 两个椭圆相交的部分既属于命题 `A`, 同时也属于命题 `B`, 即为 $A \wedge B$

如果扩展一下逻辑运算, 引入 `¬` 运算符, 则对于 $\neg(A \wedge B)$, 其真值表如下:

| A     | B     | A ∧ B | ¬(A ∧ B) |
| ----- | ----- | ----- | -------- |
| True  | True  | True  | False    |
| True  | False | False | True     |
| False | True  | False | True     |
| False | False | False | True     |

即 $\neg(A \wedge B)$ 的逻辑运算结果为 `A` 命题与 `B` 命题的逻辑 "与" 运算结果取反, 其文氏图为:

![X](./.assets/venn-diagram-for-not-and.png)

### 3.2. 逻辑 "或"

对于两个命题 `A` 和 `B`，如果其中任意一个命题为 `True`，则两个命题的 "或" 运算结果为 `True`

两个命题的 "或" 运算可写为: $A \vee B$, 表示 `A or B`

#### 3.2.1. 逻辑 "或" 真值表

基于 "或" 运算规则, 两个命题的 "或" 运算真值表如下:

| A     | B     | A ∨ B |
| ----- | ----- | ----- |
| True  | True  | True  |
| True  | False | True  |
| False | True  | True  |
| False | False | False |

从表中可以看出, 对于 `A` 和 `B` 两个命题可组合的 4 中情况中, 只需 `A` 和 `B` 其中一个为 "真" 时, $A \vee B$ 才为 "真"

#### 3.2.2. 逻辑 "或" 文氏图

对于两个命题 `A` 和 `B`, 其逻辑 "或" 运算的文氏图如下:

![X](./.assets/venn-diagram-for-or.png)

两个椭圆分别表示命题 `A` 和 `B`, 所以两个命题的 "或" 运算结果同时包含了椭圆的全部区域, 包含相交区域

如果扩展一下逻辑运算, 引入 `¬` 运算符, 则对于 $(\neg A) \vee (\neg B)$, 其真值表如下:

| A     | B     | ¬A    | ¬B    | (¬A) ∨ (¬B) |
| ----- | ----- | ----- | ----- | ----------- |
| True  | True  | False | False | False       |
| True  | False | False | True  | True        |
| False | True  | True  | False | True        |
| False | False | True  | True  | True        |

即 $(\neg A) \vee (\neg B)$ 的逻辑运算结果为 `A` 命题取反与 `B` 命题取反的逻辑 "或" 运算结果, 其文氏图为:

![X](./.assets/venn-diagram-for-not-or.png)

结合上面的 $\neg(A \wedge B)$ 逻辑运算真值表和文氏图, 可知 $\neg(A \wedge B) = (\neg A) \vee (\neg B)$, 这称为 "德摩根定律" (De Morgan's laws), 即逻辑运算的交换律和结合律

### 3.3. 逻辑 "异或"

对于两个命题 `A` 和 `B`，如果两个命题中至少有一个命题为 "真", 但又不同时为 "真"，则两个命题的 "异或" 运算结果为 "真"

两个命题的 "异或" 运算可写为: $A \oplus B$, 表示 `A exclusive or B`

#### 3.3.1. 逻辑 "异或" 真值表

基于 "异或" 运算规则, 两个命题的 "异或" 运算真值表如下:

| A     | B     | A ⊕ B |
| ----- | ----- | ----- |
| True  | True  | False |
| True  | False | True  |
| False | True  | True  |
| False | False | False |

从表中可以看出, 对于 `A` 和 `B` 两个命题可组合的 4 中情况中, 只需 `A` 和 `B` 其中一个为 "真" 时, 且不同时为 "真", $A \oplus B$ 才为 "真"

#### 3.3.2. 逻辑 "异或" 文氏图

对于两个命题 `A` 和 `B`, 其逻辑 "异或" 运算的文氏图如下:

![X](./.assets/venn-diagram-for-exclusive-or.png)

两个椭圆分别表示命题 `A` 和 `B`, 所以两个命题的 "异或" 运算结果包含了两个椭圆的不相交部分

异或也可以通过一个电路图来表示:

![X](./.assets/circuit-diagram-for-exclusive-or.png)

从上图中可以看出, 如果两个开关同时为 "开" 或 "关" 状态时, 灯泡都无法接通, 只有一个开关为 "开" 状态时, 且同时另一个开关为 "关" 状态时, 灯泡才能接通

如果扩展一下逻辑运算, 引入 `¬` 运算符, 则对于 $\neg (A \oplus B)$, 其真值表如下:

| A     | B     | A ⊕ B | ¬(A ⊕ B) |
| ----- | ----- | ----- | -------- |
| True  | True  | False | True     |
| True  | False | True  | False    |
| False | True  | True  | False    |
| False | False | False | True     |

即 $\neg(A \oplus B)$ 的逻辑运算结果为两个命题的取反, 其文氏图为:

![X](./.assets/venn-diagram-for-neg-exclusive-or.png)

可看到 $\neg(A \oplus B)$ 的结果等于 `A` 命题和 `B` 命题是否相等, 即 $\neg(A \oplus B) = (A = B)$

### 3.4. 逻辑 "相等"

对于两个命题 `A` 和 `B`，如果两个命题都为 "真" 或都为 "假", 则两个命题相等结果为 "真", 否则两个命题相等结果为 "假"

两个命题的 "相等" 运算可写为: $A = B$, 表示 `A equal to B`

#### 3.4.1. 逻辑 "相等" 真值表

基于 "相等" 运算规则, 两个命题的 "相等" 运算真值表如下:

| A     | B     | A = B |
| ----- | ----- | ----- |
| True  | True  | True  |
| True  | False | False |
| False | True  | False |
| False | False | True  |

从表中可以看出, 对于 `A` 和 `B` 两个命题可组合的 4 中情况中, 只需 `A` 和 `B` 都为 "真" 或都为 "假", $A = B$ 结果为 "真"

#### 3.4.2. 逻辑 "相等" 文氏图

对于两个命题 `A` 和 `B`, 其逻辑 "相等" 运算的文氏图如下:

![X](./.assets/venn-diagram-for-equal.png)

结合上面 $\neg(A \oplus B)$ 的结果, 可看到其结果和 `A` 与 `B` 相等的结果相同, 即 $(A = B) = \neg(A \oplus B)$

### 3.5. 逻辑 "蕴含"

对于两个命题 `A` 和 `B`，如果 `A` 命题成立, 则 `B` 命题一定也成立, 则称为 `A` 命题蕴含 `B` 命题, 例如:

`A` = "大于 10 岁的人", `B` = "大于 6 岁的人", 则 `A` 蕴含 `B`

`A` 蕴含 `B` 可写为: $A \rightarrow B$, 表示 `A equal to B`

#### 3.5.1. 逻辑 "相等" 真值表

基于 "相等" 运算规则, 两个命题的 "相等" 运算真值表如下:

| A     | B     | A = B |
| ----- | ----- | ----- |
| True  | True  | True  |
| True  | False | False |
| False | True  | False |
| False | False | True  |

从表中可以看出, 对于 `A` 和 `B` 两个命题可组合的 4 中情况中, 只需 `A` 和 `B` 都为 "真" 或都为 "假", $A = B$ 结果为 "真"

#### 3.5.2. 逻辑 "相等" 文氏图

对于两个命题 `A` 和 `B`, 其逻辑 "相等" 运算的文氏图如下:

![X](./.assets/venn-diagram-for-equal.png)

结合上面 $\neg(A \oplus B)$ 的结果, 可看到其结果和 `A` 与 `B` 相等的结果相同, 即 $(A = B) = \neg(A \oplus B)$
