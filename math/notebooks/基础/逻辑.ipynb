# 逻辑

## 1. 命题

能够判断对错的陈述句叫作命题 (Proposition), 例如如下陈述句都是命题:

- 5 + 5 = 10, 结果为 "对"
- Alice (13 岁) 的年龄在 6 岁以上, 结果为 "对"
- 5 + 5 = 11, 结果为 "错"
- Bob (6 岁) 的年龄在 6 岁以下, 结果为 "错"

命题正确时, 称该命题为 "真", 反之，命题不正确时，称该命题为 "假". "真" 称作 `true`, "假" 称作 `false`

- 5 + 5 = 10 => `true`
- Alice (13 岁) 的年龄在 6 岁以上 => `true`
- 5 + 5 = 11 => `false`
- Bob (6 岁) 的年龄在 6 岁以下 => `false`

命题要么为 `true` 要么为 `false`, 同时满足 `true` 和 `false` 的不能称为命题; 既不为 `true` 也不为 `false` 的也不能称为命题

## 2. 规则

所谓规则, 指的是一个具备自变量的命题, 当自变量的值导致命题为 `true` 或 `false` 时, 能够各自对应一个因变量, 例如一个乘车买票的规则:

- 当乘客年龄大于等于 6 岁时, 票价 100 元;
- 当乘客年龄小于 6 岁时, 票价 0 元;

其中的自变量是乘客的年龄, 因变量是票价, 因变量的计算方式是自变量是否导致命题为 `true` 或 `false`

### 2.1. 规则遗漏

一个规则应该覆盖所有可能的因变量值, 否则就会发送规则遗漏, 也就是存在至少一个自变量值, 令规则的命题既不为 `true` 也不为 `false`, 从而无法计算出因变量, 例如:

- 当乘客年龄大于 6 岁时, 票价 100 元;
- 当乘客年龄小于 6 岁时, 票价 0 元;

上述规则在 "乘客年龄等于 6 岁时", 命题既不为 `true` 也不为 `false`, 无法计算出因变量, 存在规则遗漏

### 2.2. 规则重复

所谓规则重复, 是指存在至少一个自变量, 令规则的可以同时为 `true` 和 `false`, 即规则重复, 例如:

- 当乘客年龄小于等于 6 岁时, 票价 0 元;
- 当乘客年龄大于等于 6 岁时, 票价 100 元;

当乘客年龄是 6 岁时, 上述规则产生了两个结果, 且两个结果自相矛盾, 存在规则重复

### 2.3. 完整性和排他性

在考虑规则时，确认有没有 "遗漏" 和 "重复" 是相当重要的:

- 没有 "遗漏"，即具备完整性，表示该规则在任何情况下都适用;
- 没有 "重复"，即具备排他性，表示该规则在任何情况下都不会产生矛盾;

在遇到大问题时，通常将其分解为多个小问题, 这时常用的方法就是检查规则的完整性和排他性. 即使是难以解决的大问题，也能通过这种方法转换成容易解决的小问题. 这种方法也称为 MECE（Mutually Exclusive and Collectively Exhaustive）
